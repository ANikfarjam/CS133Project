```python
import requests
from bs4 import BeautifulSoup
url ="https://www.iamexpat.de/housing/rentals/"
#for this website we are going to extract name of all the cities from German cities.csv
#then add to the ened of url and extract data
```

```python
import pandas  as pd
data = pd.read_csv("../German cities.csv")
data.head()
```

```python
city_list = data.city.to_list()
city_list[:5]
```

```python
import os
directory = "./citiesHTML/"
if not os.path.exists(directory):
    os.makedirs(directory)
for city in city_list:
    response = requests.get(url+city)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        #now lets save all the city in their respective html file
        with open(directory+city+".html", 'w') as file:
            file.write(soup.prettify())
    else:
        print("cant reach", url+city)
```

```python
import numpy as np


path = './citiesHTML/'
city_pages = os.listdir(path)
print(city_pages)



data = []

for page in city_pages:
    with open(os.path.join(path, page), 'r') as file:
        html_content = file.read()
    soup = BeautifulSoup(html_content, 'html.parser')
    
    name_tag = soup.find_all('h2', class_='article__title')
    names = [name.get_text(strip=True) if name else np.nan for name in name_tag]

    surface_tag = soup.find_all('span', class_='property__label label label--surface')
    sizes = [surface.find_next_sibling().text.strip() if surface else np.nan for surface in surface_tag]

    bedrooms_tag = soup.find_all('span', class_='property__label label label--bedrooms')
    beds = [bed.find_next_sibling('span').text.strip() if bed else np.nan for bed in bedrooms_tag]

    price_tag = soup.find_all('span', class_='property__price')
    prices = [price.get_text(strip=True) if price else np.nan for price in price_tag]

    size_modified = list(map(lambda x: re.findall(r'.{2}m', x)[0], sizes))

    # Append data to the list
    data.extend(zip(names, beds, size_modified, prices))

# Create a DataFrame
df = pd.DataFrame(data, columns=['Name', 'Beds', 'Size', 'Price'])

df
```

```python
final_df = df.drop(columns='Size')
final_df.Price = final_df.Price.str.replace('€','')
final_df
```
