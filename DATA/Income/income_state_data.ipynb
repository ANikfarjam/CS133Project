```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = "https://en.wikipedia.org/wiki/List_of_German_cities_by_GDP"
response = requests.get(url)

if response.status_code == 200:
    health_soup = BeautifulSoup(response.content, 'html.parser')
else:
    print('Couldn\'t reach', url)
```

```python
with open("income_state.html", 'w') as file:
    file.write(health_soup.prettify())
```

```python
religion_table = health_soup.find('table', class_='sortable wikitable sticky-header static-row-numbers sort-under col1left col2left')

if religion_table is not None:
    print(religion_table.prettify())
else:
    print("Table not found.")
```

```python
religion_table = health_soup.find('table', class_='sortable wikitable sticky-header static-row-numbers sort-under col1left col2left')

city_list = []
state_list = []
gdp_billions = []
gdp_per_capita = []
gdp_per_worker = []

if religion_table is not None:
    rows = religion_table.find_all('tr')
    print(rows)

    for row in rows:

        cells = row.find_all('td')

        if len(cells) == 5: # Update this number to match the number of columns in the table
            # extracting data from cells
            city = cells[0].find('a').get_text(strip=True)  
            state = cells[1].get_text(strip=True)
            gdp_bill = cells[2].get_text(strip=True)
            gdp_percap = cells[3].get_text(strip=True)
            gdp_perworker = cells[4].get_text(strip=True)

            city_list.append(city)
            state_list.append(state)
            gdp_billions.append(gdp_bill)
            gdp_per_capita.append(gdp_percap)
            gdp_per_worker.append(gdp_perworker)
else:
    print("Table not found.")
```

```python
city_income_df = pd.DataFrame({'City': city_list,
               'State': state_list,
               'GDP in Billions': gdp_billions,
               'GDP PerCapita': gdp_per_capita,
                'GDP Per Worker': gdp_per_worker})


city_income_df['GDP PerCapita'] = city_income_df['GDP PerCapita'].str.replace(',', '').astype(float)
city_income_df['GDP Per Worker'] = city_income_df['GDP Per Worker'].str.replace(',', '').astype(float)
```

```python
city_income_df.to_csv('city_income.csv', index=False)
```
