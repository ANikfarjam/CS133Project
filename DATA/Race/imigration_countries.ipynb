```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import csv
import dash
```

```python
data = pd.read_csv("region_year_imigration.csv")
```

```python
indices_to_keep = list(range(4, 14)) + list(range(16, 21)) + list(range(23, 32)) + list(range(33, 36)) + list(range(37, 40))

filtered_df = data.iloc[indices_to_keep]
```

```python
filtered_df.iloc[:, 0] = filtered_df.iloc[:, 0].str.strip()

for column in filtered_df.columns[1:]:
    if filtered_df[column].dtype == 'object': 
        filtered_df[column] = filtered_df[column].str.replace(',', '')

filtered_df = filtered_df.drop(31)
filtered_df
```
