```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/chuck/Documents/GitHub/CS133FinalProject/CS133Project/DATA/Sean data - help/Cultural/germany_state_ratings_1-10_various_categories.csv')
df.drop(['Country', 'Code'], axis = 1, inplace=True)
df.info()
df.set_index("Region")
df
```

```python
# Extract the first cell from the second column (excluding the 'Region' column)
first_cell_value = df.iloc[0, 1]  
print("First cell value from the second column:", first_cell_value)

# Generate a list of 100 values centered around that value with a normal distribution
new_values = np.random.normal(first_cell_value, scale=0.5, size=100)
print("List of 20 values centered around the first cell value with a normal distribution:")
print(new_values)
```

```python
new_dfs = []
replication_index = 1000

for i , original_row in df.iterrows():

    new_rows = []

    for j in range(replication_index):
        new_values = {}
        for column, original_value in original_row.items():
            if column == 'Region':
                continue
            new_value = np.random.normal(original_value, scale=0.5)
            new_values[column] = new_value

        new_rows.append(new_values)
    new_df = pd.DataFrame(new_rows)
    new_df['Region'] = original_row['Region']
    new_dfs.append(new_df)

merged_df = pd.concat(new_dfs, ignore_index=True)
merged_df.set_index('Region', inplace=True)
merged_df.reset_index(inplace=True)
merged_df.to_csv('resampling_germany_ratings.csv', index=False)
merged_df.info()
merged_df
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=merged_df[merged_df['Region'] == 'Schleswig-Holstein'], x='Community', label='Schleswig-Holstein')
sns.histplot(data=merged_df[merged_df['Region'] == 'Saarland'], x='Community', label='Saarland')
sns.histplot(data=merged_df[merged_df['Region'] == 'North Rhine-Westphalia'], x='Community', label='North Rhine-Westphalia')

plt.legend()
plt.show()
```
